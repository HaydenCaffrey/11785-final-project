## Combine Audio Segmentations

```python
combined_length = 2
audios_path = '/content/audios/'
training_examples = os.listdir(audios_path+'train')
val_examples = os.listdir(audios_path+'val')
test_examples = os.listdir(audios_path+'test')
num_training_examples = len(training_examples)
num_val_examples = len(val_examples)
num_test_examples = len(test_examples)
```

```python
print(num_training_examples,num_val_examples,num_test_examples)
```

```python
print(training_examples[:10])
```

```python
import random
random.seed(11785)
new_training_examples = []
new_val_examples = []
new_test_examples = []

training_idx = list(range(num_training_examples-1))
val_idx = list(range(num_val_examples-1))
test_idx = list(range(num_test_examples-1))

count_target_pos = {
    0:0,
    1:0,
}

def generate_combined_examples(raw_examples, raw_idx):
  new_examples = []
  for i, id in enumerate(raw_examples):
    combined_ids = []
    chosen_idx = random.sample(raw_idx, k=combined_length-1)
    for j in chosen_idx:
      if j < i:
        combined_ids.append(raw_examples[j][:-4])
      else:
        combined_ids.append(raw_examples[j+1][:-4])
    target_pos = random.randint(0,combined_length-1)
    count_target_pos[target_pos] += 1
    combined_ids.insert(target_pos, id[:-4])
    combined_example = {
        'audio_ids': tuple(combined_ids),
        'description': annotations_dict[id[:-4]],
        'target_id': id[:-4]
    }
    new_examples.append(combined_example)
  return new_examples


new_training_examples = generate_combined_examples(training_examples, training_idx)
new_val_examples = generate_combined_examples(val_examples, val_idx)
new_test_examples = generate_combined_examples(test_examples, test_idx)
```
